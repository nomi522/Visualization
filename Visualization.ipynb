Importing Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Loading dataset into Dataframe

```python
file_path = 'winequality-red.csv' 
df = pd.read_csv(file_path)
```

Handling missing values

```python
# Check for missing values 
missing_values = df.isnull().sum()
```

```python
#setting style for plots
sns.set_style("whitegrid")
```

Histrogram Plot for wine(Quality) vs Quantity

```python
# Histogram of wine quality
plt.figure(figsize=(8, 6))
sns.histplot(df['quality'], bins=range(2,10), kde=False, color='tomato')
plt.title('Distribution of Wine Quality')
plt.xlabel('Quality Rating')
plt.ylabel('Frequency')
plt.xticks(range(2, 10))
plt.show()
```

Barplot for wine quality

```python
average_alcohol_by_quality = df.groupby('quality')['alcohol'].mean().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(x='quality', y='alcohol', data=average_alcohol_by_quality, palette='coolwarm')
plt.title('Average Alcohol Content by Wine Quality')
plt.xlabel('Quality Rating')
plt.ylabel('Average Alcohol Content')
plt.show()
```

Heatmap to check correalation between variables

```python
# Heatmap to visualize the correlation between variables
plt.figure(figsize=(12, 10))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Wine Properties')
plt.show()
```

Scatter plot for content 

```python
# Scatter plot of alcohol content vs. wine quality
plt.figure(figsize=(10, 6))
sns.scatterplot(x='alcohol', y='quality', data=df, color='navy', alpha=0.5)
plt.title('Alcohol Content vs. Wine Quality')
plt.xlabel('Alcohol Content')
plt.ylabel('Quality Rating')
plt.show()
```
